```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
```

```python

model = load_model('CNNv2a.h5')
cap = cv2.VideoCapture(0)
l = ['Metal', 'Plastic']
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    input_frame = cv2.resize(frame, (512, 384))
    input_frame = np.expand_dims(input_frame, axis=0) / 255.0
    pred = model.predict(input_frame)
    class_index = np.argmax(pred)
    confidence = pred[0][class_index]
    label = f"Class: {l[class_index]}, con: {confidence:.2f}"
    cv2.putText(frame, label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('REFBATCH', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
